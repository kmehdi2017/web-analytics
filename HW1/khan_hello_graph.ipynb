First the libray 'networkx'  was imported that would be necessary to create a network graph. matplotlib was imported to plot
the graph:

```python
import networkx as nx
import matplotlib.pyplot as plt
```

An empty graph was created using Graph() function: 

```python
G = nx.Graph()
```

All the nodes (the names of the people in the network) were added to the graph(G) created earlier. Since the add_node_from() function takes a list as its parameter, a list of all the names (nodes) was created and passed as a parameter to 
add_node_from() function to create the nodes in the network:

```python
names = (['Andre','Beverly','Carol','Diane','Ed','Fernando','Garth','Heather','Ike','Jane'])
G.add_nodes_from(names)
```

Next, the relationships (edges) of each of the nodes (person) were arranged in Tuples, which were added in a list. Finally
the list was passed as a perameter to add_edges_from() function to add the edges to the graph:

```python
relationships = [('Andre','Beverly'),
                 ('Andre','Carol'),
                 ('Andre','Diane'),
                 ('Andre','Fernando'),
                 ('Beverly','Andre'),
                 ('Beverly','Diane'),
                 ('Beverly','Ed'),
                 ('Beverly','Garth'),
                 ('Carol','Andre'),
                 ('Carol','Diane'),
                 ('Carol','Fernando'),
                 ('Diane','Andre'),
                 ('Diane','Beverly'),
                 ('Diane','Carol'),
                 ('Diane','Ed'),
                 ('Diane','Garth'),
                 ('Diane','Fernando'),
                 ('Ed','Beverly'),
                 ('Ed','Diane'),
                 ('Ed','Garth'),
                 ('Fernando','Andre'),
                 ('Fernando','Carol'),
                 ('Fernando','Diane'),
                 ('Fernando','Garth'),
                 ('Fernando','Heather'),
                 ('Heather','Fernando'),
                 ('Heather','Garth'),
                 ('Heather','Ike'),
                 ('Ike','Heather'),
                 ('Ike','Jane'),
                 ('Jane','Ike')]
                 
G.add_edges_from(relationships) 
```

The graph was plotted after passing the attribute of nodes and edges (color, width, transparency etc.):

```python
plt.axis('off')
p = nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos=p,node_size=500, alpha=.5,node_color="r", node_shape="s", linewidths=3)
nx.draw_networkx_edges(G,pos=p,edge_color="c", width=2)
nx.draw_networkx_labels(G, pos=p, font_size=11, font_color="b" ,font_weight="bold")
plt.show()
```
